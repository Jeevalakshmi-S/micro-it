```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic data
np.random.seed(42)

# Define domains and departments
domains = ['Web Development', 'AI/ML', 'Data Science', 'Cybersecurity', 'IoT', 'Blockchain', 'Cloud Computing']
departments = ['CSE', 'ECE', 'EEE', 'IT', 'AIDS']

# Create dataframe
num_entries = 300
df = pd.DataFrame({
    'Student_ID': np.random.randint(1000, 1100, num_entries),
    'Department': np.random.choice(departments, num_entries),
    'Internship_Domain': np.random.choice(domains, num_entries),
    'Application_Date': pd.date_range(start='2024-01-01', periods=num_entries, freq='D'),
    'Status': np.random.choice(['Applied', 'Selected', 'Completed'], num_entries)
})

# Add Month column for trend analysis
df['Month'] = df['Application_Date'].dt.to_period('M')

# ---------- Plot 1: Most Popular Internship Domains ----------
plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Internship_Domain', order=df['Internship_Domain'].value_counts().index, palette='viridis')
plt.title(' Most Popular Internship Domains')
plt.xticks(rotation=45)
plt.ylabel("Number of Applications")
plt.tight_layout()
plt.show()

# ---------- Plot 2: Monthly Trends of Domains (Emerging Trends) ----------
monthly_trends = df.groupby(['Month', 'Internship_Domain']).size().unstack().fillna(0)
monthly_trends.plot(kind='line', figsize=(12,6), marker='o')
plt.title(" Monthly Trends of Internship Domains")
plt.ylabel("Number of Applications")
plt.xlabel("Month")
plt.grid(True)
plt.tight_layout()
plt.show()

# ---------- Plot 3: Department vs Domain Heatmap ----------
dept_domain = pd.crosstab(df['Department'], df['Internship_Domain'])
plt.figure(figsize=(10,6))
sns.heatmap(dept_domain, annot=True, cmap="YlGnBu", fmt='g')
plt.title(" Department vs Internship Domain Preference")
plt.ylabel("Department")
plt.xlabel("Internship Domain")
plt.tight_layout()
plt.show()

# ---------- Plot 4: Internship Status Distribution ----------
plt.figure(figsize=(6,6))
df['Status'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('pastel'), startangle=90)
plt.title(" Internship Status Distribution")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
