```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# 1Ô∏è‚É£ STEP: Generate synthetic dataset
np.random.seed(42)
num_students = 300
data = {
    'Department': np.random.choice(['CSE', 'ECE', 'EEE', 'IT', 'AIDS'], num_students),
    'GPA': np.round(np.random.uniform(6.0, 10.0, num_students), 2),
    'Completed_Projects': np.random.randint(0, 6, num_students),
    'Workshops_Attended': np.random.randint(0, 5, num_students),
    'Internship_Hours': np.random.randint(10, 200, num_students),
    'Hackathons_Participated': np.random.randint(0, 4, num_students),
    'Success': np.random.choice([0, 1], num_students, p=[0.4, 0.6])  # 1 = Success, 0 = Not
}

df = pd.DataFrame(data)

# 2Ô∏è‚É£ STEP: Data Preprocessing
# Encode 'Department'
le = LabelEncoder()
df['Department'] = le.fit_transform(df['Department'])

# Features and target
X = df.drop('Success', axis=1)
y = df['Success']

# Normalize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3Ô∏è‚É£ STEP: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# 4Ô∏è‚É£ STEP: Train Models
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(n_estimators=100),
    "Support Vector Machine": SVC(kernel='rbf')
}

# 5Ô∏è‚É£ STEP: Evaluate Models
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"\nüîç Model: {name}")
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))

# 6Ô∏è‚É£ STEP: Plot Feature Importance (for Random Forest)
importances = models['Random Forest'].feature_importances_
features = X.columns

plt.figure(figsize=(8,6))
sns.barplot(x=importances, y=features, palette='coolwarm')
plt.title("üìä Feature Importance - Random Forest")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```
